Movie Recommendation system using one hot encoding.

Part 1: One hot encoding of the Director Column

```python
import pandas as pd
import numpy as np
```

1) Loading the Dataset

```python
df=pd.read_csv('/content/drive/MyDrive/Colab Notebooks/final.csv')
print(df.head(1))
```

2) Removing unnecessary Columns

```python
#I will be recommending movies based on the director of the movie the user likes. 
#So I will remove every column except name and director.

del df['year']
del df['duration']
del df['genre']
del df['rating']
del df['type']

print(df.head())
```

3) Checking for any null values

```python
print(df.isnull().sum())
print('Number of rows: ', len(df))
```

There are no null values and the total number of rows are 1250.

4) Finding all directors

```python
directors=[]
for i in df['director']:
  directors+=i.split(',')

directors=[i.strip() for i in directors]
directors=list(set(directors))
print(len(set(directors)))
```

5) One hot Conversion

```python
one_hot=[]
for i in df['director']:
  i=[j.strip() for j in i.split(',')]
  enc=[0 for i in range(1210)]

  for direc in i:
    enc[directors.index(direc)]=1
  one_hot.append(enc)
```

6) One_hot verification

```python
for i in range(len(df['director'])):
  x1=len(df['director'][i].split(','))
  x2=list(one_hot[i]).count(1)
  
  if(x1!=x2):
    print('Not Equal!')
```

One Hot Conversion for directors is successful!

7) Converting one_hot into array

```python
one_hot=np.array(one_hot)
print(one_hot.shape)
```

8) Adding one_hot data into the dataframe

```python
for i in range(1210):
  df[directors[i]]=one_hot[:,i]
```

9) Recommending Movies (Top 10)

```python
def mod(lst):
  s=0 
  for i in lst:
    if (i<0):
      s+=(-i)
    else:
      s+=(i)
  return s
```

```python
user=input('Enter the choice of your movie: ')
user=df[df['name']==user].values[0][2:]

rel=[]
for i in df.values:
  rel.append(mod(i[2:]-user))

df['rel']=rel
df_1=df.sort_values(by='rel').head(10)

print(df_1['name'])
```

Above are the top 10 movies recommended based on the directors of the user's choice movie.
