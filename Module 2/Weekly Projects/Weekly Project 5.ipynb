**Inventory Management System**

```python
import json
import time
import pandas as pd
```

**1) Add new product in inventory.**

```python
inventory={}
```

```python
temp={}

id=int(input('Enter Product id: '))
name=input('Enter product name: ')
price=int(input('Enter price of the product: '))
category=input('Enter the product category: ')
stock=int(input('Enter number of stocks left: '))
manu=input('Enter brand manufacturer: ')

temp['product_name']=name
temp['product_price']=price
temp['product_category']=category
temp['product_stock']=stock
temp['product_manufacturer']=manu

inventory[id]=temp
```

```python
print(inventory)
```

```python
txt = ''
txt = json.dumps(inventory)

fd=open('sample_data/inventory.json','w')
fd.write(txt)
fd.close()
```

**2) Remove product from the Inventory**

```python
fd=open('sample_data/inventory.json','r')
txt=fd.read()
fd.close()

inventory_data=json.loads(txt)
print(inventory_data.keys())

rem=input('Enter Product id which has to be removed from the inventory data: ')
if (rem in inventory_data.keys()):
  del inventory_data[rem]

print(inventory_data.keys())
```

```python
txt=json.dumps(inventory_data)
fd=open('sample_data/inventory.json','w')
txt=fd.write(txt)
fd.close()
```

**3) Deduct product from inventory if purchased**

```python
fd=open('sample_data/inventory.json','r')
txt=fd.read()
fd.close()

inventory_data=json.loads(txt)
for i in inventory_data.keys():
  print('_'*25)
  print('Product ID   : ', i)
  print('Product Name : ', inventory_data[i]['product_name'])
  print('Product Price: ', inventory_data[i]['product_price'])
  print('_'*25)

def purchased(id, quantity):
  inventory_data[id]['product_stock']-=quantity

id=input('Enter the product id to be bought: ')
quantity=int(input('Enter the number of quantities to be bought: '))
purchased(id, quantity)
```

```python
txt=json.dumps(inventory_data)
fd=open('sample_data/inventory.json','w')
fd.write(txt)
fd.close()
```

**4) Generating Bill**

```python
lst=[]
lst.append((id, inventory_data[id]['product_name'],quantity,inventory_data[id]['product_price']*quantity))
df=pd.DataFrame(lst, columns=['Product ID', 'Product Name', 'Quantity', 'Price'])
print('BILL')
print('_'*50)
print(df)
```

**5) Saving bill in a csv file**

```python
df.to_csv('sample_data/lst.csv')
```
