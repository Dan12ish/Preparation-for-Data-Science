```python
import pandas as pd
```

```python
df=pd.read_csv('/content/drive/MyDrive/Colab Notebooks/country_vaccinations.csv')
print(df.head(1))
```

**1) Add month (Jan to Dec) in the dataset from data column**

```python
lst=[]
for i in df['date']:
  if (i.split('-')[1]== '01'):
    lst.append('January')
  elif (i.split('-')[1]== '02'):
    lst.append('February')
  elif (i.split('-')[1]== '03'):
    lst.append('March')
  elif (i.split('-')[1]== '04'):
    lst.append('April')
  elif (i.split('-')[1]== '05'):
    lst.append('May')
  elif (i.split('-')[1]== '06'):
    lst.append('June')
  elif (i.split('-')[1]== '07'):
    lst.append('July')
  elif (i.split('-')[1]== '08'):
    lst.append('August')
  elif (i.split('-')[1]== '09'):
    lst.append('September')
  elif (i.split('-')[1]== '10'):
    lst.append('October')
  elif (i.split('-')[1]== '11'):
    lst.append('November')
  elif (i.split('-')[1]== '12'):
    lst.append('December')

df['Month']=lst
print(df.head())
```
