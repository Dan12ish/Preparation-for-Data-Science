```python
# Program to make an attendance system which can save login and logout times
import time

attendance={}
temp={}

roll=input('Enter your roll no: ')

entry=int(input('1)Logging In? or 2)Logging Out?:' ))

if (entry==1):
  temp['Name']=input('Enter your name: ')
  temp['Logged_in_at']=time.ctime()
elif(entry==2):
  temp['Name']=input('Enter your name: ')
  temp['Logged_out_at']=time.ctime()

attendance[roll]=temp
print(attendance)
```
