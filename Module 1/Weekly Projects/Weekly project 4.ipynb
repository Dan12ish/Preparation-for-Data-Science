**Program to encrypt the files with reversing and shifting the text by 3 places** 

```python
import os
os.mkdir('sample_data/Danish')
```

**1) Creating files in bulk**

```python
for i in range(1,101):
  fd=open('sample_data/Danish/Danish_'+str(i)+'.txt', 'w')
  fd.write("Hello I am Danish and this is File no. "+str(i))
  fd.close()
```

**2) Text Encryption Function**

```python
def encrypt(text):
  t=''
  for i in text:
    t += chr(ord(i)-3)
  return t    
```

**3) Text Decryption Function**

```python
def decrypt(text):
  t=''
  for i in text:
    t +=chr(ord(i)+3)
  return t
```

**4) Encrypting of Files**

```python
for i in range(1,101):
  fd=open('sample_data/Danish/Danish_'+str(i)+'.txt','r')
  text=fd.read()
  fd.close()

  fd=open('sample_data/Danish/Danish_'+str(i)+'.txt','w')
  fd.write(encrypt(text))
  fd.close()
```

**5) Decrypting of files**

```python
for i in range(1,101):
  fd=open('sample_data/Danish/Danish_'+str(i)+'.txt','r')
  text=fd.read()
  fd.close()

  fd=open('sample_data/Danish/Danish_'+str(i)+'.txt','w')
  fd.write(decrypt(text))
  fd.close()
```
