###IMDB Celebrities Scraper with Multiprocessing

1) Importing all necessary libraries

```python
import pandas as pd
import numpy as np
import requests
from bs4 import BeautifulSoup
from tqdm import tqdm
import time
import multiprocessing
```

2) Function to scrap the data from the website.

```python
def scrap(l,n):
    
    global data
    data = []
    
    for pages in tqdm(range(1,11)):
        
        res  = requests.get(l)
        soup = BeautifulSoup(res.content, 'html.parser')
        
        for i in soup.find('div', class_='lister-list').find_all('div', class_='lister-item mode-detail'):
            
            try:
                rank = i.find('h3', class_='lister-item-header').find('span', class_='lister-item-index unbold text-primary').text.split('.')[0].strip()
            except:
                rank = np.nan
    
            try:
                actor_id = i.find_all('a')[1].get('href').split('/')[-1].strip()
            except:
                actor_id = np.nan
    
            try:
                name = i.find_all('a')[1].text.strip()
            except:
                name = np.nan
    
            try:
                img_link = i.find('a').find('img').get('src').strip()
            except:
                img_link = np.nan
    
            try:
                exp = i.find('p').text.split(' ')[25].strip()
            except:
                exp = np.nan
    
            try:
                work = i.find('p').find('a').text.strip() + ' (' + i.find('p').find('a').get('href').split('/')[-2].strip() + ')'
            except:
                work = np.nan
    
            data.append([rank, actor_id, name, img_link, exp, work])
    
        l = 'https://www.imdb.com' + soup.find('a', class_='lister-page-next next-page').get('href').strip()

    global df
    df = pd.DataFrame(data, columns = ['Rank', 'Actor_ID', 'Name', 'Img_link', 'Experince', 'Work'])
    df.to_csv('/content/drive/MyDrive/Colab Notebooks/' + n + '.csv', index = False)
```

3) Using multiprocessing for scrapping the data 

```python
p1 = multiprocessing.Process(target = scrap , args = ('https://www.imdb.com/search/name/?match_all=true&ref_=nv_cel_m', '1-10', ))
p2 = multiprocessing.Process(target = scrap , args = ('https://www.imdb.com/search/name/?match_all=true&start=501&ref_=rlm', '11-20', ))

p1.start()
p2.start()

p1.join()
p2.join()
```
