###  Github Profile Scraper for all the Followers

1) Importing all the required Libraries

```python
import numpy as np
import pandas as pd
import time
from tqdm import tqdm
import requests
from bs4 import BeautifulSoup
```

2) Sending the Request to the initial page for data

```python
link = 'https://github.com/AshishJangra27?tab=followers'
res  = requests.get(link)
soup = BeautifulSoup(res.content, 'html.parser')
```

3) Finding the number of pages of followers we need to iterate through

```python
followers = int(soup.find_all('span', class_ = 'text-bold color-fg-default')[0].text.strip())

pages = int(followers/50)

if (followers%50 != 0):
            pages+=1
```

4) Scrapping the data

```python
data = []


for j in range(1, pages+1):                          # To iterate through all the follower pages
    
    res  = requests.get(link)
    soup = BeautifulSoup(res.content, 'html.parser')
    
    for i in tqdm(soup.find_all('div', class_ = 'd-table table-fixed col-12 width-full py-4 border-bottom color-border-muted')):
        
        res  = requests.get('https://github.com' + i.find('a').get('href'))
        soup = BeautifulSoup(res.content, 'html.parser')
    
        try:
            username = soup.find('span', class_ = 'p-nickname vcard-username d-block').text.strip()
        except:
            username = np.nan
    
        try:
            name = soup.find('span', class_ = 'p-name vcard-fullname d-block overflow-hidden').text.strip()
        except:
            name = np.nan
    
        try:
            profile_image_link = soup.find('img', class_ = 'avatar avatar-user width-full border color-bg-default').get('src')
        except:
            profile_image_link = np.nan
    
        try:
            location = soup.find('span', class_ = 'p-label').text.strip()
        except:
            location = np.nan
    
        try:
            followers = soup.find_all('span', class_ = 'text-bold color-fg-default')[0].text.strip()
        except:
            followers = np.nan
    
        try:
            following = soup.find_all('span', class_ = 'text-bold color-fg-default')[1].text.strip()
        except:
            following = np.nan
    
        try:
            no_of_repositories = soup.find_all('a', class_= 'UnderlineNav-item js-responsive-underlinenav-item js-selected-navigation-item')[0].find('span').text.strip()
        except:
            no_of_repositories = np.nan
    
        try:
            no_of_projects = soup.find_all('a', class_= 'UnderlineNav-item js-responsive-underlinenav-item js-selected-navigation-item')[1].find('span').text.strip()
        except:
            no_of_projects = np.nan
    
        try:
            no_of_stars = soup.find_all('a', class_= 'UnderlineNav-item js-responsive-underlinenav-item js-selected-navigation-item')[3].find('span').text.strip()
        except:
            no_of_stars = np.nan
    
        data.append([username, name, profile_image_link, location, followers, following, no_of_repositories, no_of_projects, no_of_stars])
    
    res  = requests.get(link)                                    # To get back to the follower list page as inside the loop goes to the individual follower's profile 
    soup = BeautifulSoup(res.content, 'html.parser')
    link = soup.find('a', rel = 'nofollow').get('href').strip()  # Getting the link to the next follower's page
    
```

5) Saving the Data as a dataframe

```python
df = pd.DataFrame(data, columns = ['Username', 'Name', 'Profile_Image_Link', 'Location', 'Followers', 'Following', 'no_of_repositories', 'no_of Projects', 'no_of_stars'])
```
